# Transcription — Pathway C: Manual

For languages not supported by Whisper or Meta MMS — or any community that prefers not to use automated tools regardless of coverage.

Manual transcription by fluent speakers produces the highest-quality data. It is the approach used by Te Hiku Media for Māori (310+ hours), and it is the correct first step for building a model in any language that lacks one. The transcribed recordings you produce here are the training data that makes future automation possible on community terms.


## 1. Workflow

1. Open each WAV file in Audacity (or any audio player)
2. Listen and type the transcript directly into `metadata/metadata_template.csv` in the `transcript` column
3. Preserve the orthography exactly — do not simplify or substitute ASCII approximations
4. Have a fluent speaker or knowledge keeper review each transcript
5. Set `knowledge_keeper_reviewed = true` in the metadata once confirmed

## 2. Orthography and Encoding

All files in this project use **UTF-8 encoding**, which supports the full Unicode range needed for Indigenous language orthographies. Write characters directly — no escaping or transliteration needed.

Examples of characters that must be preserved exactly:

| Language | Characters |
|---|---|
| Navajo (Diné bizaad) | Ł, á, ą, ą́, é, ę, ę́, í, į, į́, ó, ǫ, ǫ́, ń |
| Māori | ā, ē, ī, ō, ū (macrons) |
| Mi'kmaq | Various diacritics depending on orthography used |

If your transcription tool saves in a different encoding, convert before importing:
```bash
iconv -f latin1 -t utf-8 input.csv > output_utf8.csv
```

## 3. Keyboard Input

[Keyman](https://keyman.com/) provides free, community-developed keyboards for many Indigenous languages including Navajo, Cree (syllabics), Cherokee, and others. It is cross-platform and integrates with standard text editors and spreadsheet tools.

For Navajo specifically, the [Navajo keyboard for Keyman](https://keyman.com/keyboards/navajo) covers all required characters.

## 4. Validate Your CSV

Once transcription is complete, run this cell to check for common issues before moving on.

````python
import pandas as pd

METADATA_CSV = "../metadata/metadata_template.csv"

df = pd.read_csv(METADATA_CSV, dtype=str).fillna("")
issues = []

empty_transcripts = df[df["transcript"].str.strip() == ""]
if len(empty_transcripts):
    issues.append(f"{len(empty_transcripts)} rows with empty transcript: {empty_transcripts['file_id'].tolist()}")

not_reviewed = df[
    (df["exclude_from_training"].str.lower() != "true") &
    (df["knowledge_keeper_reviewed"].str.lower() != "true")
]
if len(not_reviewed):
    issues.append(f"{len(not_reviewed)} rows not yet knowledge-keeper reviewed: {not_reviewed['file_id'].tolist()}")

if issues:
    print("Issues to resolve before export:")
    for issue in issues:
        print(f"  - {issue}")
else:
    print(f"All {len(df)} rows look good — ready to export.")

df[["file_id", "transcript", "knowledge_keeper_reviewed", "consent_tier"]]
````
