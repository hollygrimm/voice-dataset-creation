## Part 4: From Governance to Code

In the real pipeline, every governance decision you just made eventually becomes a line of code. The [export notebook](../notebooks/05_export_ljspeech.ipynb) uses this filter:

```python
eligible = df[
    (df["exclude_from_training"].str.lower() != "true") &
    (df["consent_tier"].str.lower().isin(["open", "community"])) &
    (df["transcript"].str.strip() != "")
]
```

Let's apply that filter to the repository's sample metadata to see the principle in action.

````python
# Load the repository's actual sample metadata
sample = pd.read_csv("../metadata/metadata_template.csv", dtype=str).fillna("")

print("All recordings in the archive:")
print(sample[["file_id", "consent_tier", "exclude_from_training", "cultural_protocol"]].to_string(index=False))

print("\n" + "=" * 60)

# Apply the same filter used in the export notebook
eligible = sample[
    (sample["exclude_from_training"].str.lower() != "true") &
    (sample["consent_tier"].str.lower().isin(["open", "community"])) &
    (sample["transcript"].str.strip() != "")
]

excluded = sample[~sample.index.isin(eligible.index)]

print(f"\nEligible for training: {len(eligible)} of {len(sample)}")
print(f"Excluded from training: {len(excluded)} of {len(sample)}")

if len(excluded) > 0:
    print("\nExcluded recordings:")
    for _, row in excluded.iterrows():
        reason = row.get("exclude_reason", "") or f"consent_tier={row['consent_tier']}"
        print(f"  {row['file_id']}: {reason}")
        if row["cultural_protocol"]:
            print(f"    Protocol: {row['cultural_protocol']}")
````
